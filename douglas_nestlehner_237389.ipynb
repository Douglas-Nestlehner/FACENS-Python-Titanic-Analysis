# RMS Titanic

O RMS Titanic foi um navio de passageiros britânico operado pela White Star Line e construído pelos estaleiros da Harland and Wolff em Belfast. Foi a segunda embarcação da Classe Olympic de transatlânticos depois do RMS Olympic e seguido pelo HMHS Britannic. Projetado pelos engenheiros navais Alexander Carlisle e Thomas Andrews, sua construção começou em março de 1909 e ele foi lançado ao mar em maio de 1911. O Titanic foi pensado para ser o navio mais luxuoso e mais seguro de sua época, gerando lendas que era supostamente "inafundável".

A embarcação partiu em sua viagem inaugural de Southampton para Nova Iorque em 10 de abril de 1912, no caminho passando em Cherbourg-Octeville na França e por Queenstown na Irlanda. Ele colidiu com um iceberg às 23h40min do dia 14 de abril e afundou na madrugada do dia seguinte matando mais de 1500 pessoas, sendo um dos maiores desastres marítimos em tempos de paz de toda a história. Seu naufrágio destacou vários pontos fracos de seu projeto, deficiências nos procedimentos de evacuação de emergência e falhas nas regulamentações marítimas da época. Comissões de inquérito foram instauradas nos Estados Unidos e no Reino Unido, levando a mudanças nas leis internacionais de navegação que permanecem em vigor mais de um século depois.

Os destroços do Titanic foram procurados por décadas até serem encontrados em 1985 por uma equipe liderada por Robert Ballard. Ele se encontra a 3843 m de profundidade e a 650 km ao sudeste de Terra Nova no Canadá. Sua história e naufrágio permaneceram no imaginário popular durante décadas, levando a produção de vários livros e filmes a seu respeito, mais notavelmente o filme Titanic de 1997. Até hoje o Titanic permanece como um dos navios mais famosos da história, com seus destroços atraindo várias expedições de exploração ao longo dos anos.

## Importando os Dados

Iniciamos a análise carregando as bibliotecas necessárias, e realizando a leitura do arquivo 'titanic_data.csv' pela função ``` read_csv() ``` da biblioteca ```pandas```.

Optamos por definir todas as informações como tipo texto (```dtype = str ```), para em seguida transformar as variáveis numéricas para seus respectivos tipos, evitando transformações desnecessárias.  

````python
# Carregando bibliotecas e lendo os dados
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

tit = pd.read_csv('/content/titanic_data.csv', sep = ',', dtype = str)
tit.head(5)
````

### Tratamentos

Para começar analisar a base, observamos informações basicas como: quantiade de Linhas e colunas:



````python
# Apresentando qtd colunas e linhas
tit.info()
````

Também removemos colunas que julgamos não essencias para a análise. Para isso, utilizamos a função ```drop()```

````python
# Removendo colunas que nao serao utilizadas
columnDrop = ['SibSp', 'Parch', 'Ticket']
tit = tit.drop(columns=columnDrop)

# Renomeando colunas
columnNames = ['IdPassageiro', 'Sobreviveu', 'Classe', 'Nome', 'Sexo', 'Idade', 'Tarifa', 'Cabine', 'Embarque']
tit.columns = columnNames
tit.head(4)
````

Em seguida, utilizando a função ```astype()```, para realizar a transformação de variáveis numéricas para o tipo ```float```.

````python
# Tratando variaveis numéricas (Age, SibSp, Parch e Fare)
columnNumeric = ['Idade', 'Tarifa']
for column in columnNumeric:
    tit[column] = tit[column].astype(float)

#tit.head(4)
````

Antes de observar medidas e indicadores, é importante observar se a base apresenta informações vazias ```NA's```, pois podem afetar diretamente os possiveis indicadores.

Para isso, utilizamos as funções ```isna()``` (indica se a observação é NA ou não) e ```sum()``` (responsavel por somar a quantidade)

````python
# Verificando valores nulos em colunas numericas
qtdNA = tit[columnNumeric].isna().sum()
print(qtdNA)

print(round(qtdNA[0]/len(tit)*100,2))
````

Temos que 19,87% das informações de idade não estão preenchidas !

Para que essas observações não sejam desconsideradas do estudo, podemos utilizar métodos de input de dados, como exemplo:

*   Substituir valores vazios por medidas da variável:
  * Média;
  * Mediana;
  * Moda;
  * entre outras.
*   Substituir valores que condizem com as caracteristicas das demais variáveis (métodos de predição):
  * Regressão Simples, Multipla, Polonomial, LASSO, ...
  * Arvores;
  * KNN;
  * Modelos estatísitcos em geral;


Nesse caso, iremos preecher as informações utilizando a mediana, mas ressaltamos que poderia ser realizado um estudo mais especifico,
realizando um input mais preciso. Principalmente devido ao fato que 20% da base apresenta valores ```NA``` e que com o input utilizando a mediana provavelmente tera impacto nos indicadores relacionados a idade.

````python
# Substituindo NAs pela mediana
tit['Idade'] = tit['Idade'].fillna(tit['Idade'].median())
````

Realizamos o tratamento de renomear alguns fatores de variáveis, utilizando a função ```replace()```.

````python
# Renomeando observações
tit['Sobreviveu'].replace('0', 'Não', inplace = True) # inplace = True permite a transformação direta no df
tit['Sobreviveu'].replace('1', 'Sim', inplace = True)

tit['Sexo'].replace('female', 'Mulher', inplace = True)
tit['Sexo'].replace('male', 'Homem', inplace = True)
````

## Análise Exploratória

Uma das principais formas de explorar/descrever os dados de uma base, é observar, em variáveis numéricas, medidas como: média, mediana, min, max, etc. O que permite ter uma ideia das distribuições das variáveis, suas principais características, e até mesmo possíveis inconsistências.

As medidas citadas poderiam ser calculadas pelas bibliotecas ```pandas``` e ```numpy```. Mas nesse caso, optamos por construir um boxplot das variáveis, o qual carrega as principais medidas de uma forma resumida e de fácil visualização.

````python
# Criando a figura
fig = plt.figure(figsize =(8, 4))

# Definindo os eixos
ax = fig.add_axes([0, 0, 1, 1])

# Criando o boxplot
bp = ax.boxplot([tit['Idade'], tit['Tarifa']], patch_artist=True, labels=['Idade', 'Tarifa'])

# Definir limites para o eixo x e y
#ax.set_xlim(0, 0)
ax.set_ylim(0, 250)

# Adiconando medidas
ageMean = tit['Idade'].mean()
tarifaMean = tit['Tarifa'].mean()
ax.text(0.15, 0.95, f'Média Idade: {ageMean:.2f}', ha='center', va='center', transform=ax.transAxes, fontsize = 9)
ax.text(0.15, 0.90, f'Média Tarifa: {tarifaMean:.2f}', ha='center', va='center', transform=ax.transAxes, fontsize = 9)

# Definindo cores
colors = ['dodgerblue', 'cornflowerblue']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# Definindo cores das extremidades da caixa
for whisker in bp['whiskers']:
    whisker.set(color='black', linewidth=1.5, linestyle=':')

# Personalizando a mediana
for median in bp['medians']:
    median.set(color='red', linewidth=1.5)

# Personalizando outliers
for flier in bp['fliers']:
  flier.set(marker='D', color='#e7298a', alpha=0.5)

# Definido o titulo
plt.title("Boxplot Idade e Tarifa")

# Exibindo o gráfico
plt.show()
````

Realizando uma análise rapida sobre o box-plot, é possivel destacar:

*   **Variável Idade**:
  * População relativamente jovem;
  * Outliers são observados, porém nenhum caso discrepante que não condiz com a realidade, exemplo idade = 240, o que indicaria uma inconsistencia da base;
* **Variável Tarifa:**
  * Destaque para casos com tarifa muito maior do que a média observada, casos podem ser um fator de comparação em realação a variável resposta (Sobrevivencia).






Além de observar variáveis numéricas, também podemos observar variáveis categóricas, que permitem trazer informações sobre proporções, contagens, moda, entre outras medidas.  

A seguir, apresentamos algumas informações sobre as variáveis categóricas:

````python
# Obtendo qtd pessoas por sexo
qtdSex = tit['Sexo'].value_counts().sort_values(ascending=True)
print(qtdSex)

# Criando a figura
plt.figure(figsize =(8, 4))
plt.barh(qtdSex.index, qtdSex.values, color=['pink', 'cornflowerblue'])

plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.title('Contagem de Homens e Mulheres')

# Adicionando as qtd de cada barra
for index, value in enumerate(qtdSex.values):
    plt.text(value, index, str(value), va='center', ha='right')

plt.show()
````

Observando que a proporção de homens no navio é bem maior do que mulheres, e levando essa informação em conta, tambem podemos ter como interesse observar a proporção de sobreviventes em relação ao sexo.

Para isso, utilizamos a função ```groupby()```, que permite agrupar os dados com base em uma ou mais colunas, ```size()``` responsavel por trazer as contagens, e ```unstack()``` por estruturar o resultado removendo os indices.

````python
# Obtendo a qtd de sobreviventes por sexo
QtdSurvSex = tit.groupby(['Sexo', 'Sobreviveu']).size().unstack()
print(QtdSurvSex)

# Criando grafico de barras empilhadas
QtdSurvSex.plot(kind='bar', stacked = True)

plt.xlabel('Sexo')
plt.ylabel('Quantidade')
plt.title('Quantidade de Sobreviventes e Não Sobreviventes por Sexo')

plt.show()
````

Além das diversas informações apresentadas até então, outras diversas questões e indicadores podem ser observados.

Pensando nesses contexto, definimos uma função chamada ```countSurv()```, que permite trazer a quantidade de sobreviventes conforme as características que definirmos.

Exemplo:


*  A Quantidade de pessoas de Mulheres, que estavamo nas classes 2 ou 3, e que tinham idade menor do que 30.



````python
# Definindo uma funcao que filtre informações especificas e retorne a qtd
def countSurv(tit, ageStart, ageFinish, sex, classe):
    # Filtrando os dados
    filter = tit[
        (tit['Idade'] >= ageStart) &
        (tit['Idade'] <= ageFinish) &
        (tit['Sexo'].isin( list(sex) ) ) &
        (tit['Classe'].isin(list(classe)))
        ]

    # Obtendo qtd sobreviventes
    qtdSurv = filter['Sobreviveu'].eq('Sim').sum()
    qtdNSurv = filter['Sobreviveu'].eq('Não').sum()

    total = len(filter)
    percSurv = str(round((qtdSurv/total)*100, 2)) + '%'
    percNSurv = str(round((qtdNSurv/total)*100, 2)) + '%'

    return {
        'Sobreviventes': qtdSurv,
        'Não Sobreviventes': qtdNSurv,
        'Porcentagem Sobreviventes': percSurv,
        'Porcentagem Não Sobreviventes': percNSurv
    }

# Aplicando em um exemplo
surv = countSurv(tit, ageStart=0, ageFinish=30, sex=['Mulher'], classe=['3','2'])
print(f'Informações: {surv}')
````
